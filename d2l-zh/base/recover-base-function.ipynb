```python
# 函数是数学的概念
```

```python
# 定义一个函数
def func(x, y, z):
    return x + y -z
```

```python
# 调用函数
result = func(1, 2, 3)
print(result)
```

```python
# 无返回值函数
def no_return_func(x, y, z):
    print(x +y + z)
```

```python
# 调用函数
no_return_func(1, 2, 3)
```

```python
# 什么是内置函数： 官方在Builtins模块下定义好的函数叫做内置函数
# 案例，定义一个函数输出开多少次方的结果
def seq(x, y):
    return x ** (1 / y)
print(seq(4, 2))
# py中开方运算得出的结果也是浮点数
```
