```python
# 跟随时间打印进度
import time
for i in range(101):
    print(f"\r进度: [{i:3}%] {'█' * (i // 2)}{' ' * (50 - i // 2)}", end='', flush=True)
    time.sleep(0.05)
print()  # 最后换行
```

```python
# 捕获打印内容为字符串：通过将标准输出 sys.stdout临时重定向到 io.StringIO对象，可以捕获 print函数输出的内容，用于调试或记录
from io import StringIO
output = StringIO()
print("Hello, StringIO!", file=output)
captured_text = output.getvalue()
print(f"捕获到的内容：{captured_text}")
output.close()
```

```python
# 通过自定义函数，你可以无限扩展 print的功能。
# 自动添加时间戳：创建一个自定义打印函数，为每次输出自动加上时间戳，便于日志记录

from datetime import datetime
def my_print(*args, **kwargs):
    timestamp = datetime.now().strftime("[%Y-%m-%d %H:%M:%S]")
    print(timestamp, *args, **kwargs)
my_print("用户登录成功！")
```

```python
# 实现条件打印：在调试时，你可能只想在特定条件下（如调试模式开启）才打印信息。可以封装一个条件打印函数
DEBUG = True
def debug_print(*args, **kwargs):
    if DEBUG:
        print("[DEBUG]", *args, **kwargs)
debug_print("这只在DEBUG为True时显示。")
```
